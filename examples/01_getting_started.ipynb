# Getting Started with ValueInvestPy

This notebook demonstrates the core features of the `ValueInvestPy` library. We will walk through fetching data, performing fundamental analysis, screening for stocks, visualizing data, and backtesting a simple portfolio.

```python
# Add the project root to the Python path to allow importing the library
import sys
import os
sys.path.insert(0, os.path.abspath('..'))

# Import the necessary functions from our library
from src.valueinvestpy.data_fetch import fetch_stock_data, fetch_financials
from src.valueinvestpy.analysis import calculate_ratios, calculate_graham_number
from src.valueinvestpy.screening import screen_stocks
from src.valueinvestpy.visuals import plot_price_history
from src.valueinvestpy.portfolio import build_portfolio
from src.valueinvestpy.backtest import run_backtest

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Step 1: Fetching Stock Data

```python
# Let's fetch historical data and financials for Microsoft (MSFT)
ticker = 'MSFT'
stock_data = fetch_stock_data(ticker, period='5y')
financials = fetch_financials(ticker)

print(f"Successfully fetched data for {ticker}.")
print("Historical Prices:")
print(stock_data.head())
print("
Income Statement:")
print(financials['income_statement'].head())
```

## Step 2: Fundamental Analysis

```python
# Calculate key financial ratios and the Graham Number
ratios = calculate_ratios(ticker)
graham_num = calculate_graham_number(ticker)

print(f"Key Ratios for {ticker}:")
for key, value in ratios.items():
    print(f"  {key.replace('_', ' ').title()}: {value:.2f if isinstance(value, float) else value}")

print(f"
Graham Number for {ticker}: ${graham_num:.2f}")
```

## Step 3: Screening for Value Stocks

```python
# Let's screen a few tech stocks based on value criteria
tech_stocks = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'META']
criteria = {
    'pe_ratio_max': 40.0,          # Max P/E of 40
    'return_on_equity_min': 0.20   # Min ROE of 20%
}

print("Screening for stocks with P/E < 40 and ROE > 20%...")
screened_results = screen_stocks(tech_stocks, criteria)
print("
Screening Results:")
print(screened_results)
```

## Step 4: Visualizing Data

```python
# Plot the historical price of our stock
fig = plot_price_history(ticker, period='10y')
plt.show()
```

## Step 5: Building and Backtesting a Portfolio

```python
# Create a simple portfolio with equal weights
portfolio_tickers = ['AAPL', 'MSFT', 'GOOG']
my_portfolio = build_portfolio(portfolio_tickers)

print(f"Portfolio created with tickers: {my_portfolio.tickers}")
print(f"Portfolio weights: {my_portfolio.weights}")

# Run a buy-and-hold backtest
performance = run_backtest(my_portfolio)

print("
Backtest Performance Summary:")
for key, value in performance.items():
    if key != 'cumulative_returns':
        print(f"  {key.replace('_', ' ').title()}: {value}")
```

```python
# Plot the cumulative returns from the backtest
if 'cumulative_returns' in performance:
    plt.figure(figsize=(12, 7))
    performance['cumulative_returns'].plot()
    plt.title('Portfolio Cumulative Returns')
    plt.ylabel('Cumulative Returns')
    plt.grid(True)
    plt.show()
```
